```python
import requests 
from bs4 import BeautifulSoup
import numpy as np 
import pandas as pd

final = pd.DataFrame()

for i in range(1,178):
  name = []
  image = []
  description = []
  cuisine = []
  course = []
  diet = []
  prep_time = []
  serving = []
  ingredients = []
  instructions = []
  print('################################### Scraping Page {} ########################################'.format(i))
  print('')
  html = requests.get('https://www.archanaskitchen.com/recipes/indian/page-{}'.format(i)).text
  soup = BeautifulSoup(html)   
  counter = 0
    
  for item in soup.find_all('a',{"itemprop":'url'}):
    print("scraping item {}".format(counter))
    url = 'https://www.archanaskitchen.com/' + item['href']
    html = requests.get(url).text

    soup = BeautifulSoup(html)
    try:
      # fetch name of recipe
      text = soup.find('h1',class_='recipe-title')
      name.append(text.text)

      # fetch image url

      img = soup.find('img',class_='img-fluid img-thumbnail')
      image.append("https://www.archanaskitchen.com" + img['src'])
    except:
      continue
        
      # fetch description
    try:
      description_div = soup.find('div',class_='row recipedescription')
      description_text = description_div.find_all('p')[1]
      description.append(description_text.text)
    except:
      description.append(np.nan)
        
    # fetch  type
    try:
      cuisine_type = soup.find('span',{'itemprop':'recipeCuisine'})
      cuisine.append(cuisine_type.text)
    except:
      cuisine.append(np.nan)
        
    # fetch course type
    try:
      course_type = soup.find('div',{'class':'col-12 course'})
      course.append(course_type.find('span',{'itemprop':'keywords'}).text)
    except:
      course.append(np.nan)
        
        # fetch diet type
    try:
      diet_type = soup.find('div',{'class':'col-12 diet'})
      diet.append(diet_type.find('span',{'itemprop':'keywords'}).text)
    except:
      diet.append(np.nan)
            
    # fetch prep time
    try:
      item = soup.find('div',{'class':'row RecipeServesTime'})
      prep_time.append(item.find_all('div',{'class':'col-md-2 col-3'})[2].find('span',{'itemprop':'totalTime'}).text.strip())
    except:
      prep_time.append(np.nan)
 
    try:
      serving_type = soup.find('div',{'class':'row RecipeServesTime'})
      serving.append(serving_type.find_all('div',{'class':'col-md-2 col-3'})[3].find('span',{'itemprop':'recipeYield'}).text.strip())
    except:
      serving.append(np.nan)
            
        # fetch ingredients
    try:
      item = soup.find('div',{'class':'row recipeingredientsanddirections'})
      ingredients.append(item.find('ul',{'class':'list-unstyled'}).text)
    except:
      ingredients.append(np.nan)
            
        # fetch instructions
    try:
      item = soup.find('div',{'class':'col-md-8 col-12 recipeinstructions'})
      instructions.append(item.find('ol').text)
    except:
      instructions.append(np.nan)
        
            
        
    counter = counter + 1

  print('')
  print(str(counter) + " items in page " + str(i) )

  temp_df = pd.DataFrame() 
  temp_df['name'] = name
  temp_df['image_url'] = image
  temp_df['description'] = description
  temp_df['cuisine'] = cuisine
  temp_df['course'] = course
  temp_df['diet'] = diet
  temp_df['prep_time'] = prep_time
  temp_df['ingredients'] = ingredients
  temp_df['instructions'] = instructions
    
  final = final.append(temp_df,ignore_index=True)
```

```python
final.to_csv('cuisines.csv',index=False)
```

```python
import pandas as pd
import os
from tqdm import tqdm
import urllib.request

df = final

message = []
counter = 1
for name in tqdm(df['image_url']):
  try:
    urllib.request.urlretrieve(name, os.path.join("data",str(counter) + "." + name.split('/')[-1]))
    message.append(1)
  except:
    message.append(0)
  counter = counter + 1
```

```python
df['image_available'] = message
```
